```python
import csv
import numpy as np
```

```python
samples = []
with open('./data/driving_log.csv') as csvfile:
    reader = csv.reader(csvfile)
    # next(reader, None)  # skip the headers
    for line in reader:
        samples.append(line)
```

```python
samples.head()
```

```python
angles = []
for sample in samples:
    angle_center = float(sample[3])
    angles.append(angle_center)
```

```python
angles
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(angles,bins=50)
plt.xlim(-0.5,0.5)
plt.show()
```

```python
x = range(len(angles))
plt.scatter(x,angles)
plt.plot(x,angles)
plt.xlim(0,1000)
plt.ylim(-0.5,0.5)
plt.show()
```
